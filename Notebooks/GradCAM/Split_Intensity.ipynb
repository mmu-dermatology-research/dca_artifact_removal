```python
# 3. Get the inside mask
inside_mask = ImageOps.invert(outside_mask)
plt.imshow(inside_mask, cmap = 'gray')
```

```python
# returns average pixel brightness
def brightness_avg(img, mask):
    stat = ImageStat.Stat(img, mask)
    return stat.mean[0]

# return RMS pixel brightness
def brightness_rms(img, mask):
    stat = ImageStat.Stat(img, mask)
    return stat.rms[0]
```

```python
outside_rms = brightness_rms(image, outside_mask)
print('outside rms: ' + str(outside_rms))

outside_brightness = brightness_avg(image, outside_mask)
print('outside brightness: ' + str(outside_brightness))

inside_rms = brightness_rms(image, inside_mask)
print('inside rms: ' + str(inside_rms))

inside_brightness = brightness_avg(image, inside_mask)
print('inside brightness: ' + str(inside_brightness))
```
