```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dca_size = ["Small", "Medium", "Large", "Oth"]

binary_dca_baseline_accuracy = [61, 63, 55, 60]
realistic_dca_baseline_accuracy = [60, 64, 60, 58]
clean_baseline_accuracy = [59, 57, 51, 58]
clean_ns_accuracy = [59, 62, 64, 58]
clean_telea_accuracy = [59, 62, 65, 58]

plt.figure(0)
plt.plot(dca_size, clean_baseline_accuracy, label='Baseline', linestyle= 'dashed')

plt.plot(dca_size, clean_ns_accuracy, label='NS Inpainting')
plt.plot(dca_size, clean_telea_accuracy, label='Telea Inpainting')

plt.plot(dca_size, binary_dca_baseline_accuracy, label='Binary DCA')
plt.plot(dca_size, realistic_dca_baseline_accuracy, label='Realistic DCA')

plt.title('Accuracy Comparison Between Inpainted Images \n and Models trained on Augmented Images')
plt.xlabel('DCA Size')
plt.ylabel('Accuracy (%)')

plt.legend()
plt.grid()
plt.show()
```

```python
clean_baseline_f1 = [0.68, 0.68, 0.67, 0.67]
clean_ns_f1 = [0.68, 0.70, 0.70, 0.67]
clean_telea_f1 = [0.68, 0.70, 0.71, 0.67]
binary_dca_baseline_f1 = [0.70, 0.72, 0.68, 0.67]
realistic_dca_baseline_f1 = [0.68, 0.68, 0.50, 0.66]



plt.figure(1)
plt.plot(dca_size, clean_baseline_f1, label='Baseline', linestyle= 'dashed')

plt.plot(dca_size, clean_ns_f1, label='NS Inpainting')
plt.plot(dca_size, clean_telea_f1, label='Telea Inpainting')

plt.plot(dca_size, binary_dca_baseline_f1, label='Binary DCA')
plt.plot(dca_size, realistic_dca_baseline_f1, label='Realistic DCA')

plt.title('F1 Score Comparison Between Inpainted Images \n and Models trained on Augmented Images')
plt.xlabel('DCA Size')
plt.ylabel('F1 Score')

plt.legend()
plt.grid()
plt.show()
```
